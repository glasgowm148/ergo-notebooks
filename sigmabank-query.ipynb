# Tidy

The UTXO model tracks the total balance on each transaction in nanoErgs. Some small fixes to make it more readable.

```python
# Finding difference
df['BalanceChange (ERG)'] = df.value.diff()
df['HeightChange (h)'] = df.creationHeight.diff()
# nanoErg -> Erg
df['value'] = df['value'] / 1000000000
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'] / 1000000000
# Track this from geneis block ?
df['HeightChange (h)'] = df['HeightChange (h)'] * 2
df['HeightChange (h)'] = df['HeightChange (h)'] / 60
# Reversing polarity
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'] * -1 
df['HeightChange (h)'] = df['HeightChange (h)'] * -1 
df['BalanceChange (ERG)'] = df['BalanceChange (ERG)'].shift(-1)
df['HeightChange (h)'] = df['HeightChange (h)'].shift(-1)

#df
```

# Tokens

```python
# Pulling banks SigUSD assets

df3 = df['assets'].apply(pd.Series)

#df3

df_sigUSD = df3[0].apply(pd.Series)
df_sigUSD.drop(columns=['tokenId','index','decimals','type', 
                 ], inplace=True)

df_sigUSD = df_sigUSD.rename(columns={'amount': 'AmountUSD'})
#df_sigUSD
```

```python
df_sigRSV = df3[1].apply(pd.Series)
df_sigRSV.drop(columns=['tokenId','index','decimals','type', 
                 ], inplace=True)
df_sigRSV = df_sigRSV.rename(columns={'amount': 'AmountRSV'})

#df_sigRSV
```

```python
# Merging back into original frame and tidying
df = pd.concat([df, df_sigRSV,df_sigUSD], axis=1)
df.drop(columns=['assets', 'name'], inplace=True)
df = df.rename(columns={'value': 'Final Balance'})
```
